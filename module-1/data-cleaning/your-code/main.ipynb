# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
db_connection_str = 'mysql+pymysql://guest:relational@relational.fit.cvut.cz:3306/stats'
db_connection = create_engine(db_connection_str)
```

#### 4. Import the users table.

```python
query_users = 'SELECT * FROM users'
df_users = pd.read_sql(query_users, con=db_connection)
df_users.head()
```

#### 5. Rename Id column to userId.

```python
df_users = df_users.rename(columns={'Id': 'userID'})
df_users.head()
```

#### 6. Import the posts table. 

```python
query_posts = 'SELECT * FROM posts'
df_posts = pd.read_sql(query_posts, con=db_connection)
df_posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
df_posts = df_posts.rename(columns={'Id': 'postID','OwnerUserId':'userID'})
df_posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
#columns selected
users_columns=['userID', 'Reputation', 'Views', 'UpVotes', 'DownVotes']
posts_columns= ['postID', 'Score', 'userID', 'ViewCount', 'CommentCount']

#new data frames
new_df_users = df_users[users_columns]
new_df_posts = df_posts[posts_columns]

#print df
display(new_df_users.head())
display(new_df_posts.head())
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(new_df_posts,new_df_users,on='userID',how='inner')
merged_df.head()
merged_df.shape
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
#we can know how many missing values and where they are with this method.
merged_df.info(memory_usage='deep')
merged_df.isnull().sum()
```
